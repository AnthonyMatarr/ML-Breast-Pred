```python
import sys

sys.path.append("../")
from src.data_utils import get_feature_lists
from src.config import BASE_PATH

import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
```

```python
data_dir = (
    BASE_PATH / "data" / "raw" / "cleaned_combined" / "NSQIP_mast_base.parquet"
)

df = pd.read_parquet(data_dir)
```

```python
##Imported func from src
feature_lists = get_feature_lists(df)
binary_cols = feature_lists["binary_cols"]
numerical_cols = feature_lists["numerical_cols"]
nominal_cols = feature_lists["nominal_cols"]
ordinal_cols = feature_lists["ordinal_cols"]
```

```python
numerical_cols
```

```python
num_cols = []
for col in numerical_cols:
    if "CPT" in col or "PODIAG" in col:
        continue
    num_cols.append(col)

df[num_cols].median()
```

Numerical

```python
for feature in numerical_cols:
    plt.figure(figsize=(12, 6))

    # Histogram with KDE
    plt.subplot(1, 2, 1)
    sns.histplot(df[feature], kde=True, bins=50, color="blue")
    plt.title(f"Histogram of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Count")

    # Q-Q Plot
    plt.subplot(1, 2, 2)
    stats.probplot(df[feature], dist="norm", plot=plt)
    plt.title(f"Q-Q Plot of {feature}")

    # Adjust layout and display
    plt.tight_layout()
    plt.show()
```
